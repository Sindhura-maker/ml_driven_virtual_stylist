```python
# Imports
import pandas as pd
import os
import random
from PIL import Image, ImageDraw

# Create folder for images
os.makedirs("small_fashion_dataset/images", exist_ok=True)

# Define sample values
genders = ["Men", "Women"]
masterCategories = ["Apparel", "Footwear", "Accessories"]
subCategories = ["Topwear", "Bottomwear", "Shoes", "Bags"]
articleTypes = ["Tshirt", "Jeans", "Sneakers", "Handbag"]
baseColours = ["Red", "Blue", "Black", "White", "Green"]
seasons = ["Summer", "Winter", "Monsoon"]
usages = ["Casual", "Office", "Party", "Date"]
budgets = ["Low", "Medium", "High"]
style_personalities = ["minimal", "sporty", "street", "classic", "party"]
fits = ["Loose", "Regular", "Slim"]
comfort_styles = ["Comfort-first", "Style-first"]

# Generate synthetic dataset
data = []
for i in range(1, 51):
    item = {
        "id": i,
        "gender": random.choice(genders),
        "masterCategory": random.choice(masterCategories),
        "subCategory": random.choice(subCategories),
        "articleType": random.choice(articleTypes),
        "baseColour": random.choice(baseColours),
        "season": random.choice(seasons),
        "usage": random.choice(usages),
        "price": random.choice([199,299,399,499,599]),
        "budget": random.choice(budgets),
        "style_personality": random.choice(style_personalities),
        "fit": random.choice(fits),
        "comfort_style": random.choice(comfort_styles),
    }
    data.append(item)

    # Dummy image
    img = Image.new('RGB', (128, 128), color=random.choice(["red","blue","green","black","white"]))
    draw = ImageDraw.Draw(img)
    draw.text((10,50), f"ID:{i}", fill="yellow")
    img.save(f"small_fashion_dataset/images/{i}.jpg")

# Save CSV
df = pd.DataFrame(data)
df.to_csv("small_fashion_dataset/styles_small.csv", index=False)
print("Dataset created with 50 items")
df.head()
```

```python
def detect_body_shape(shoulder, bust, waist, hip):
    if bust > hip and bust > waist:
        return "Inverted Triangle"
    if hip > bust and hip > waist:
        return "Pear"
    if abs(bust - hip) < 5:
        return "Hourglass"
    if waist > bust:
        return "Apple"
    return "Rectangle"

# Example
body_shape = detect_body_shape(38,36,30,38)
print("Detected Body Shape:", body_shape)
```

```python
# Placeholder function (replace with actual image-based detection if needed)
def get_skin_tone(user_skin="light"):
    tones = {"light":[255,224,189], "medium":[220,180,160], "dark":[120,60,40]}
    return tones.get(user_skin, [220,180,160])

skin_tone = get_skin_tone("medium")
print("Detected skin tone:", skin_tone)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import NearestNeighbors
import numpy as np

# Encode categorical columns
categorical_cols = ['gender','masterCategory','subCategory','articleType','baseColour','season','usage','budget','style_personality','fit','comfort_style']
encoders = {}
for col in categorical_cols:
    encoders[col] = LabelEncoder()
    df[col] = encoders[col].fit_transform(df[col])

# Prepare feature matrix
feature_cols = categorical_cols + ["price"]
X = df[feature_cols]

# Train KNN
knn = NearestNeighbors(n_neighbors=5, metric='euclidean')
knn.fit(X)

# Recommendation function
def recommend_outfit(user_vector):
    distances, indices = knn.kneighbors([user_vector])
    return df.iloc[indices[0]]

# Example user vector (random)
user_vec = X.iloc[0].values
recommend_outfit(user_vec)
```

```python
def suggest_accessories(body_shape, occasion):
    acc = []
    if body_shape == "Pear":
        acc += ["Statement earrings", "Structured bags"]
    elif body_shape == "Inverted Triangle":
        acc += ["Chunky bracelets", "A-line bags"]

    if occasion == "party":
        acc += ["Clutch bag", "Heels"]
    elif occasion == "office":
        acc += ["Minimal watch", "Leather belt"]

    return list(set(acc))

suggest_accessories("Pear", "party")
```

```python
def makeup_suggestions(skin_tone_family, occasion):
    base = "Warm beige" if skin_tone_family == "warm" else "Cool ivory"
    lips = "Nude peach" if occasion == "office" else "Red berry"
    eyes = "Brown matte" if occasion == "office" else "Gold shimmer"
    return {"Base": base, "Lipstick": lips, "Eye Look": eyes}

makeup_suggestions("warm", "party")
```

```python
def color_palette(skin_tone_family):
    if skin_tone_family=="warm":
        return {"best_colors":["Red","Orange","Olive"], "avoid_colors":["Blue","Purple"]}
    else:
        return {"best_colors":["Blue","Purple","Magenta"], "avoid_colors":["Orange","Yellow"]}

color_palette("warm")
```
